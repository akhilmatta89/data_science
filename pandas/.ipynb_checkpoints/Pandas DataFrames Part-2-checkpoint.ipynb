```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.arange(20).reshape(5,4), ['A','B','C','D','E'],['W','X','Y','Z'])
```

```python
df
```

```python
#Very important feature of pandas is it has ability to perform conditional selection using bracket notation
```

```python
df[df>0]   #Boolean values will convert to NAN
```

```python
df['W']>0
```

```python
df[df['W']>0]  #in this case the total false row will be gone
# will not effect original dataframe
```

```python
df
```

```python
df[df['W']>0][['Y','X']]
```

```python
from numpy.random import randn
np.random.seed(101)
```

```python
df=pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df
```

```python
#use & or | symbol for multiple conditions
df[(df['W']>0) & (df['Y']<0)]
```

```python
df[(df['W']>0) | (df['Y']<0)]
```

```python
df
```

```python
df.reset_index()
#changes indexes
```
